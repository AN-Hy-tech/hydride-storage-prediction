```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

def save_subplots(image_path1, image_path2, save_directory, output_filename='combined_predvsactual.png'):
    # ایجاد یک figure جدید با زمینه سفید
    fig, axs = plt.subplots(1, 2, figsize=(15, 10), facecolor='white')
    
    # خواندن و نمایش تصویر اول
    img1 = mpimg.imread(image_path1)
    axs[0].imshow(img1)
    axs[0].axis('off')
    
    # خواندن و نمایش تصویر دوم
    img2 = mpimg.imread(image_path2)
    axs[1].imshow(img2)
    axs[1].axis('off')
    
    # تنظیم زمینه سفید برای figure
    fig.patch.set_facecolor('white')
    
    # مسیر کامل برای ذخیره
    save_path = os.path.join(save_directory, output_filename)
    
    # ذخیره figure به عنوان فایل PNG
    plt.savefig(save_path, bbox_inches='tight', dpi=300)
    plt.close(fig)  # بستن figure برای جلوگیری از نمایش
    
    print(f"Plot saved successfully to: {save_path}")

# مسیرهای تصاویر و دایرکتوری ذخیره
image_path1 = r"D:\Hydride_Machine_learning_project\Machine_Model\reports\figures\default_benchmark_analysis\pred_vs_actual_best.png"
image_path2 = r"D:\Hydride_Machine_learning_project\Machine_Model\reports\figures\default_benchmark_analysis\pred_vs_actual_ensemble.png"
save_directory = r"D:\Hydride_Machine_learning_project\Machine_Model\reports\figures\default_benchmark_analysis"

# فراخوانی تابع برای ذخیره
save_subplots(image_path1, image_path2, save_directory)
```
